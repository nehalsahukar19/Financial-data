```python
#Major Indices
major_indices = pd.read_html("https://finance.yahoo.com/world-indices?guccounter=1")[0]
```

```python
major_indices.head()
```

```python
ticker_list=major_indices['Symbol'].str.replace("^","").str.lower().to_list()
```

```python
len(ticker_list)
```

```python
df=yf.download(ticker_list,period="1d",start="2020-01-13",end="2021-03-10")#date of covid starts and it ends
```

```python
df.columns
```

```python
df.head()
```

```python
adj_close = df.dropna(thresh=10,axis=1)['Adj Close']
```

```python
adj_close.head()
```

```python
adj_close.describe().T
```

```python
adj_close.plot(figsize=(12,12),subplots=True)
```

```python
adj_close=adj_close.drop('XAX',axis=1)
```

```python
adj_close.plot(figsize=(12,20),subplots=True)
```

```python
dji=adj_close['DJI']
```

```python
dji.mean()
```

```python
dji
```

```python
dji.shift(1)
```

```python
dji_per_change = dji / (dji.shift(1) - 1)
```

```python
dji_per_change.plot(figsize=(12,6))
```

```python
dji_log_returns_shift = np.log(dji / dji.shift(1))
```

```python
dji_log_returns_shift.plot(figsize=(12,6))
```

```python
dji.hist(bins=50, figsize=(12,12))
```
